```python
df_hierarchy=pd.DataFrame(columns=["level","place","geneo"])
len(df_hierarchy)
```

```python
def find_by_key(data,c, target):
    c=c+1
    for key, value in data.items():
        if isinstance(value,dict):
            yield from find_by_key(value,c, target)
        elif key == target:
            yield c,value
```

```python
l0=0
x0="root"
dlevel={5:1,8:2,11:3,14:4}
for l,x in find_by_key(hierarchy,0,"title"):
    df_hierarchy=df_hierarchy.append({"level":dlevel[l],"place":x,"geneo":{}},ignore_index=True)
```

```python
df_hierarchy[['previous']]=df_hierarchy[["level"]].shift(1) # vorherige level
df_hierarchy['previous'].iloc[0]=0
df_hierarchy["ID"]=df_hierarchy.index
df_hierarchy.head(10)
```

```python
class Gen:
    glevel= dict_x = collections.defaultdict(list)
    
    def __init__(self):
        ob = collections.defaultdict(list)
        self.ob={1:0}
```

```python
def parent(x):
    y=int(x["ID"])
    l=int(x["level"])
    p=int(x["previous"])
    if y<1: 
        g={1:0}
    elif l>p and p>0:
        g={l:y-1}
        gg.ob[l]=y-1
    elif l==p:
        g={l:gg.ob[l]}
    elif l<p:
        g={l:gg.ob[l]}
    return(g)
```

```python
gg=Gen()
df_hierarchy["geneo"]=df_hierarchy.apply(lambda x3: parent(x3),axis=1)
df_hierarchy.head(5)
```

```python
def top(x):
    l=x["level"]
    ref=x["geneo"][l]
    y=df_hierarchy.iloc[ref]
    pla=y["place"]
    return(pla)    
```

```python
df_hierarchy["hoehle"]=df_hierarchy.apply(lambda x4: top(x4),axis=1)
df_hierarchy.head(5)
```
