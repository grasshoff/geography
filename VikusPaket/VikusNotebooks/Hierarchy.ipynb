```python
def find_by_key(data,c, target):
    c=c+1
    for key, value in data.items():
        if isinstance(value,dict):
            yield from find_by_key(value,c, target)
        elif key == target:
            yield c,value
```

```python
for l,x in find_by_key(hierarchy,0,"title"):
#    print("level=",l,"x=",x)
    df_hierarchy=df_hierarchy.append({"level":dlevel[l],"place":x,"geneo":{}},ignore_index=True)
```
